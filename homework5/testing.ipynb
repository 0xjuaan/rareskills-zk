```python
from py_ecc.bn128 import G1, G2, pairing, add, multiply, eq, neg
```

```python

"""
To generate the points: 

The whole idea is that e(aG1, bG2) = e(G1, G2)^(ab)

This means that e(aG1, bG2) * e(cG1, dG2) = e(G1, G2)^(ab + cd)

-> pairing(a1, b2) = pairing(G1, G2)^210 = p(24G1, 5G2) * p(6G1, 7G2) * p(12G1, 4G2)

You can see how 24*5 + 6*7 + 12*4 = 210

Equation is:
pairing(a1, b2) == pairing(alpha1, beta2) * pairing(x1, gamma2) * pairing(c1, delta2)
"""
# LHS
a1 = multiply(G1, 2 * 3 * 7)
b2 = multiply(G2, 5)

# RHS
alpha1 = multiply(G1,2 * 3 * 4)
beta2 = multiply(G2, 5)

x1 = multiply(G1, 6)
gamma2 = multiply(G2, 7)

c1 = multiply(G1, 12)
delta2 = multiply(G2, 4)

## Assert the eqn
assert(pairing(b2, a1) == pairing(beta2, alpha1) * pairing(gamma2, x1) * pairing(delta2, c1))

## Evaluate the identity element of G12
print(pairing(b2, neg(a1))*  pairing(beta2, alpha1) * pairing(gamma2, x1) * pairing(delta2, c1))
```
